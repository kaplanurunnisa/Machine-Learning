```python
from PIL import Image
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
import pandas as pd
from sklearn import tree
from sklearn import neural_network


def rgb_to_int(r,g,b):
    
    return (r<<16) + (g<<8) + b

def read_data(filename):
    x = []
    y = []
    back_color = rgb_to_int(255,255,255)
    
    image = Image.open(filename)
    width,height = image.size
    pixels = image.load()
 
    for i in range(width):
        for j in range(height):
            #print(pixels[i,j])
            r,g,b = pixels[i,j]
            color = rgb_to_int(r,g,b)
            
            if (color != back_color):
                x.append([i,j])
                y.append(color)
    return x,y
```

*2. Create two datasets based on the 2 images from step 1, using the code give below. Analyze and comment this code.*

```python
x1, y1 = read_data("firstpicc.png")
y1
x2, y2 = read_data("secondpicc.png")
y2
print(y1)
print(y2)
```

*3. Split the first dataset into a training set and a test set (using 70% for training and 30% for validation)*
